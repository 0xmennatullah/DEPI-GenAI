```python
class Perceptron:
    def __init__(self, lr=0.01, epochs=10):
        self.lr = lr 
        self.epochs = epochs
        self.weights = None
        self.bias = None

    #activation func, loss, packprop
    def activation(self, x):
        return np.where(x>=0, 1, 0)
        #step function! first backprop funtion in perceptron

    def fit(self, X, y):
        n_samples, n_features = X.shape()
        self.weights = np.zeros(n_features)
        self.bias = 0
        for i in range(self.epochs): # whole data x times
            print(f"epoch {i} : Weights {self.weights}, Bias {self.bias}")
            for idx, x_i in enumerate(X): # each row in data
                # trainning => weighted sum
                linear_out = np.dot(X, self.weights) + self.bias
                prediction = self.activation(linear_out)

                #update rule => SGD
                update = self.lr * (y[idx] - prediction)
                self.weights += update * x_i
                self.bias += update
       
    def predict(self, X):
        linear_out = np.dot(X, self.weights) + self.bias
        prediction = self.activation(linear_out)
        return prediction
```

```python
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0, 1, 1, 0])
# XOR gate
```

```python
# object mn el perceptron
p = Perceptron(lr=0.1, epochs=10)
p.fit(X, y)
```

```python
p.predict(X)
```
