```python
import torch.nn as nn 
import torch.optim as optim
from torchvision import datasets, transforms #cv in pytorch, pyaudio/pyvision/etc
from torch.utils.data import DataLoader
```

```python
transform = transforms.Compose([
    transforms.ToTensor()
])
```

```python
train_dataset = datasets.MNIST(
    root='./data',
    train=True,
    transform = transform,
    download=True
)
```

```python
test_dataset = datasets.MNIST(
    root='./data',
    train=False,
    transform = transform,
    download=True
)
```

```python
train_loader = DataLoader(train_dataset, batch_size = 64, shuffle=True )
test_loader = DataLoader(test_dataset, batch_size=64)
```

```python
class NeuralNet(nn.Module):
    def __init__(self):
        super().__init__()

        self.fc1 = nn.Linear(28*28, 128)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = x.view(-1, 28*28) #don't care bout first dim, set def => flatten the data
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
```

```python
model = NeuralNet()
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters())
```

```python
for epochs in range(5):
    for images, labels in train_loader:
        output = model(images)
        loss = loss_fn(output, labels)
        optimizer.zero_grad()
        loss.backward() # calc new grad
        optimizer.step() #optimizer edit ur params
    print(f"epoch: {epochs+1} || loss = {loss.item():.4f}")
```

```python
correct = 0
total = 0

with torch.no_grad():
    for image, label in test_loader:
        output = model(image)
        _,predict = torch.max(output,1)
        total+= label.size(0)
        correct += (predict)
```

```python
%pip install tensorflow
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
(x_train, y_train), (x_test, y_test) = tf.akeras.datasets.mnist.load_data()
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
model = models.Sequential([
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(10)
])
```

```python
loss_fn = tf.keras.losses.SparseCategoricalCrossEntropy(from_logits=True)

optimizer = tf.keras.optimizers.Adam()
```

```python
model.compile(optimizer, loss_fn)
model.fit(x_train, y_train)
```

```python
output = model.predict(x_test)
pred_idx = tf.argmax(output, axis=1)
tf.keras.metrics.Accuracy(y_test, pred_idx)
```
